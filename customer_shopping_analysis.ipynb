```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.offline import iplot,plot
```

```python
df=pd.read_csv('/content/customer_shopping_data.csv')
```

```python
df
```

```python
#dataset information
df.info()
```

```python
df.isna().sum() # pandas is used to check for missing values (NaN) in a DataFrame and to count the number of missing values in each column.
```

```python
#No Nan values
df.duplicated().sum()
```

No duplicates
This dataset is clean from dupliacates and NaN values


#Preprocessing the data

```python
#Droping the invoice_no column because we will not use it in analysis.and inplace=True meaning original object is directly modified without creating a new one
df.drop(columns=['invoice_no'],inplace=True)
```

```python
df
```

```python
# Setting the customer_id as index for this dataset
df.set_index('customer_id', inplace=True)
```

```python
df
```

```python
#changing the datatype of the invoice_date column from object to dateTime
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
```

```python
df.info()
```

```python
#Adding (year,month) columns.Maybe we will use them in a analysis
df['year']=df['invoice_date'].dt.year
df['month']=df['invoice_date'].dt.month
```

```python
df.head()
```

```python
df.info()
```

#Analysis & Visualization


```python
#gender count
gender_count=df['gender'].value_counts()
gender_count
```

```python
sns.countplot(x='gender',data=df,palette=['pink','skyblue'])
plt.title=('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Another plot for gender distribution using plotly just for practice
iplot(px.bar(gender_count,
             title='Gender distribution',
             text_auto=True,
             labels=dict(index='gender', value='count'),
             color_discrete_sequence=[['#FF69B4', '#87CEEB']]).update_xaxes(type='category'))
```

As we can see the Female customers are more than the male customers


```python
#The df.describe method in pandas is used to generate descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset’s distribution, excluding NaN values. This method can be applied to both numeric and object data type
df.describe()
```

```python
#age count
age_count=df['age'].value_counts().sort_index()
age_count
```

```python
max_age = df.nlargest(1,'age')
max_age
```

```python
min_age=df.nsmallest(1,'age')
min_age
```

```python
#getting mean age by gender
gender_age=df.groupby('gender')['age'].mean().reset_index()
```

```python
gender_age
```

```python
sns.barplot(x= 'gender', y = 'age', data = gender_age, palette=['pink','skyblue'])
plt.title=('Mean age by gender')
plt.xlabel('Gender')
plt.ylabel('Mean Age')
plt.show()
```

```python
# another plot using plotly just for practice
iplot(px.bar(gender_age, x='gender', y='age', color='gender',
             color_discrete_sequence=['pink', 'skyblue'],
             labels={'gender': 'Gender', 'age': 'Mean Age'},
             title='Mean age by gender').update_xaxes(categoryorder='total ascending'))
```

It seems like they are so near from each other

```python
#category_count
category_count=df['category'].value_counts()
category_count
```

```python
sns.barplot(x=category_count.index, y=category_count.values)
plt.title=('Count of Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Clothing is the most Category that the customers buy

```python
## another plot using plotly just for practice
iplot(px.bar(x=category_count.index, y=category_count.values,
             labels={'x': 'Category', 'y': 'Count'},
             title='Count of Each Category',
             color=category_count.index,
             color_discrete_sequence=px.colors.qualitative.Set3).update_layout(xaxis_tickangle=-45))
```

```python
#insights between category and gender
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x=df['category'], hue=df['gender'], palette=['pink','skyblue'])
plt.title=('Category count for each gender')
```

We can notice that females buy more than males in every category

```python
# another plot using plotly just for practice
gender_category_count = df.groupby(['gender', 'category']).size().reset_index(name= 'count')
iplot(px.bar(gender_category_count, x='category', y='count', color='gender',
             barmode='group', labels={'category': 'Category', 'count': 'Count'},
             title='Category Count for Each Gender',
             color_discrete_sequence=['pink','skyblue']))
```
